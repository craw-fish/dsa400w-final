```python
def consolidate_names(df : pd.DataFrame, col: str, block_by = None, threshold : int = 85, scorer=fuzz.ratio):
    names = pd.DataFrame({'val': df[col]})
    results_map = {}
    
    # block by elements
    names['block'] = block_by
    
    for _, block_df in names.groupby('block'):
        # unique names ordered by frequency
        uniques = block_df['val'].value_counts().index.to_list()
        
        mapping = {}
        used = set()
        
        # iterate through unique names (in order of frequency)
        for canonical in uniques:
            # skip if name already matched (variant of a canonical name)
            if canonical in used:
                continue
            # identify similar strings (fuzzy matching)
            matches = process.extract(
                canonical,
                uniques,
                scorer=scorer,
                score_cutoff=threshold
            )
            # get name values from match arrays
            match_names = [match[0] for match in matches]
            # construct dict mapping variant names to canonical name (highest freq)
            for name in match_names:
                mapping[name] = canonical
                used.add(name)
                # print any changed items
                # if name != canonical:
                #     print(f'{name} -> {canonical}')
            
        results_map.update(mapping)
                
    results = df[col].map(results_map)
    
    return results, results_map
```

```python
# Match city names between properties and address points
def match_names(col : pd.Series, choices, scorer=fuzz.token_set_ratio, threshold=90):
    uniques = list(col.unique())
    if isinstance(choices, pd.Series):
        choices = choices.unique()
    mapping = {}
    
    for name in uniques:
        match = process.extractOne(
            name,
            choices,
            scorer=scorer,
            score_cutoff=threshold
        )
        if match:
            match_name = match[0]
            mapping[name] = match_name
        else:
            mapping[name] = name
    
    results = col.map(mapping)
    
    return results, mapping
```

### Additional matching

```python
def fuzzy_match_addr(left_df : pd.DataFrame, right_df : pd.DataFrame,
                     left_on = 'GEO_MERGE_KEY', right_on = 'GEO_MERGE_KEY',
                     left_zip = 'PROPERTYZIP', right_zip = 'ZIP_CODE', 
                     left_num = 'PROPERTYHOUSENUM', right_num = 'ADDR_NUM',
                     scorer = fuzz.WRatio, threshold = 0):
    
    results = []
    
    # limit comparisons to same zip and house number to avoid false matches (+ improve performance)
    for (zipcode, housenum), left_subset in left_df.groupby([left_zip, left_num]):
        subset_results = left_subset.copy()
        
        zipcode = int(zipcode)
        housenum = int(housenum)
        right_subset = right_df[
            (right_df[right_zip].astype(int) == zipcode) & 
            (right_df[right_num].astype(int) == housenum)
        ]
        
        # no matches
        if len(right_subset) == 0:
            continue
        
        left_vals = left_subset[left_on].values
        right_vals = right_subset[right_on].values
        
        # calculate fuzzy match matrix
        scores = process.cdist(
            left_vals,
            right_vals,
            scorer=scorer
        )
        
        match_idx = scores.argmax(axis=1)
        match_scores = scores.max(axis=1)
        
        subset_results['FUZZY_KEY'] = right_vals[match_idx]
        subset_results['FUZZY_SCORE'] = match_scores
        
        # drop results that did not meet threshold (0 in score matrix)
        subset_results.loc[subset_results['FUZZY_SCORE'] < threshold, 'FUZZY_KEY'] = None
        
        results.append(subset_results)
    
    return pd.concat(results)
        
```

```python
def fuzzy_match_name(col : pd.Series, search_space, scorer = fuzz.WRatio, threshold = 0):    
    # order by most frequent
    uniques = col.value_counts().index
    choices = np.array(search_space)
    results = pd.DataFrame(columns=['NAME', 'MATCH', 'SCORE']).set_index('NAME')
    
    for name in uniques:
        match = process.extractOne(
            name,
            choices,
            scorer=scorer
        )
        
        match_name = match[0]
        match_score = match[1]
        
        if match_score >= threshold:
            results.loc[name, 'MATCH'] = match_name
            results.loc[name, 'SCORE'] = match_score
        # else:
        #     results.loc[name] = None
    
    return results
        
```
